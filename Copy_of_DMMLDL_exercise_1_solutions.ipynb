## Task 1

a) Probability that the new test declares that a randomly selected processor does not have a defect: 0.35

> D = defect

> ND = no defect

> PT = passes test (no defect)

> NT = doesn't pass test (defect)

> P(D) = 0.8

> P(PT|ND) = 0.95 

> P(PT|D) = 0.2 


Law of total probability:

> P (PT) = P(PT|D) x P(D) + P(PT|ND) x P(ND) 

> P (PT) = 0.2 x 0.8 + 0.95 x 0.2

> P (PT) = 0.16 + 0.19

> P (PT) = 0.35

...

b) Probability that a processor that passed the new test has a defect: 0.457


> D = defect

> ND = no defect

> PT = passes test

> NT = doesn't pass test


Bayes Theorem:

>P (D|PT)  = P (PT|D) P(D)  /  P (PT)

>P (D|PT)  =  0.2 x 0.8 / 0.35 

>P (D|PT)  = 0.457


...

c) Probability that a processor that did not pass the test does not have a defect: 0.015

>D = defect

>ND = no defect

>PT = passes test (no defect)

>NT = doesn't pass test (defect)

>P(ND) = 1 - P(D) = 0.2


Law of total probability:

>P (NT) = P(NT|D) x P(D) + P(NT|ND) x P(ND) 

>P (NT) = 0.8 x 0.8 + 0.05 x 0.2

>P (NT) = 0.65


Bayes Theorem:

>P (ND|NT)  = P (NT|ND) P(ND)  /  P (NT)

>P (ND|NT)  = 0.05 x 0.2 /  0.65

>P (ND|NT)  =  0.01538

```
x = np.linspace(-5, 5, 5000)
mu = 0
sigma = 1

y_pdf = stats.norm.pdf(x, mu, sigma) # the normal pdf
y_cdf = stats.norm.cdf(x, mu, sigma) # the normal cdf

plt.plot(x, y_pdf, label='pdf')
plt.plot(x, y_cdf, label='cdf')
plt.legend();
```
